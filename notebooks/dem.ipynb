```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
```

```python
dem_path = "../data/processed/dem/dem_guaiba_clip.tif"

with rasterio.open(dem_path) as src:
    dem = src.read(1)          # primeira banda (elevação)
    transform = src.transform
    crs = src.crs
    nodata = src.nodata
```

```python
print("CRS:", crs)
print("Shape:", dem.shape)
print("NoData:", nodata)
print("Min:", np.nanmin(dem))
print("Max:", np.nanmax(dem))
```

```python
dem = dem.astype(float)

if nodata is not None:
    dem[dem == nodata] = np.nan
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(dem, cmap="terrain")
plt.colorbar(label="Elevação (m)")
plt.title("DEM – Bacia do Guaíba")
plt.axis("off")
plt.show()
```
